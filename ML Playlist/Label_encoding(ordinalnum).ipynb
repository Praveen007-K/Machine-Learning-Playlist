```python
import pandas as pd
import datetime
```

```python
#list of 20 dates diff from today

today = datetime.datetime.today()
df_date_list = [today - datetime.timedelta(days = x) for x in range (20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
#extract weekday name
df['date_of_week'] = df['day'].dt.day_name()
df.head()
```

```python
# Create ranking dict
day_map = {'Monday':1,
          'Tuesday':2,
          'Wednesday':3,
          'Thursday':4,
          'Friday':5,
          'Saturday':6,
          'Sunday':7}

df['day_ordinal'] = df.date_of_week.map(day_map)
df.head()
```
