```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.savefig('Test.png')
```

```python
plt.plot(x,x*x,'g*--') #ro,r*,etc
```

```python
plt.plot(x,x*x,'r',linestyle='dashed',linewidth=4,markersize=1)
```

```python
#subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(3,1,1)
plt.plot(x,y_sin)
plt.title("Sine")

plt.subplot(3,1,3)
plt.plot(x,y_cos,'r--')
plt.title("Cosine")

plt.show()
```

```python
x = [2,7,10]
y = [3,9,6]

x1 = [3,5,9]
y1 = [1,3,1]

plt.bar(x,y,align='center')#??
plt.bar(x1,y1,color='g')
plt.title("Bar graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
a = np.array([2,4,5,7,12,43,3,65,21,34,6,87,54,32,11,37,75,74,36])
plt.hist(a,bins=20)
plt.title("Histogram")
plt.show()
```

```python
data = [np.random.normal(0,std,100)for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True);
```

```python
labels = 'Chicken','Egg','Fish','Vegetables'
sizes = [240,180,200,120]
colors = ['yellow','lightgreen','pink','lightblue'] #lightcoral,lightsyblue
explode = (0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
